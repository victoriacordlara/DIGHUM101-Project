#### 3.2 Tokenize and Remove StopWords

```python
# Converting to lower
for k, v in song_dic.items():
    song_dic[k] = v.lower()

# Removing Punctuation
from string import punctuation
for char in punctuation:
    for k, v in song_dic.items():
        song_dic[k] = v.replace(char, "")

# Tokenizing
token_dic = {}
for k,v in song_dic.items():
    token_dic[k] = v.split()

# Removing StopWords
import nltk
from nltk.corpus import stopwords

stop_sp = stopwords.words("spanish")
stop_sp = stop_sp + ["pa", "to", "yo", "tu", "tú", "te", "si", "má", "ey", "pal", "yeh",
                    "uh", "ah"] # these are abreviated versions of stopwords
stop_en = stopwords.words("english") + ["im", "ill", "like", "yah", "oh", "ooh",
                                        "youre", "youll","yeah", "hey"] # there are some english phrases / songs

for k,v in token_dic.items():
    token_dic[k] = [word for word in v if word not in stop_sp]
    token_dic[k] = [word for word in token_dic[k] if word not in stop_en]

token_dic
```

### 4. Exploratory Data Analysis

Now that the is data in a legible and easily manipulated format, we can start creating visuals that will display the patterns that are present. This is done for each year first and then for the entirety of the data.

#### 4.1 Word Frequency Analysis and WordClouds

```python
# Word Counter per year

# First need to have all the words of the years in one list
tok18 = sum(list(songs_df[songs_df["Year"] == 2018]["Tokens"]),[])
tok19 = sum(list(songs_df[songs_df["Year"] == 2019]["Tokens"]),[])
tok20 = sum(list(songs_df[songs_df["Year"] == 2020]["Tokens"]),[])
tok21 = sum(list(songs_df[songs_df["Year"] == 2021]["Tokens"]),[])
tok22 = sum(list(songs_df[songs_df["Year"] == 2022]["Tokens"]),[])

# Count word frequencies
from collections import Counter
freq18 = Counter(tok18).most_common()
freq19 = Counter(tok19).most_common()
freq20 = Counter(tok20).most_common()
freq21 = Counter(tok21).most_common()
freq22 = Counter(tok22).most_common()

# Create dataframes of top 50 most common to plot later
freq18_df = pd.DataFrame(data = freq18, columns = ["Word", "Frequency"]).head(50)
freq19_df = pd.DataFrame(data = freq19, columns = ["Word", "Frequency"]).head(50)
freq20_df = pd.DataFrame(data = freq20, columns = ["Word", "Frequency"]).head(50)
freq21_df = pd.DataFrame(data = freq21, columns = ["Word", "Frequency"]).head(50)
freq22_df = pd.DataFrame(data = freq22, columns = ["Word", "Frequency"]).head(50)

# Word Clouds
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

def wordcloud_maker(DataFrame):
    text = DataFrame.Word.values
    wordcloud = WordCloud(
        width = 3000,
        height = 2000,
        background_color = 'black',
        stopwords = STOPWORDS).generate(str(text))
    fig = plt.figure(
    figsize = (8, 6),
    facecolor = 'k',
    edgecolor = 'k')
    plt.imshow(wordcloud, interpolation = 'bilinear')
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
    return wordcloud
```
